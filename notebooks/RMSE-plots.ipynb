```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
MLP_ADA = pd.read_parquet("../data/RMSPE/MLP-ADA.parquet")
MLP_AXS = pd.read_parquet("../data/RMSPE/MLP-AXS.parquet")
MLP_BTC = pd.read_parquet("../data/RMSPE/MLP-BTC.parquet")
MLP_DOGE = pd.read_parquet("../data/RMSPE/MLP-DOGE.parquet")
MLP_NEAR = pd.read_parquet("../data/RMSPE/MLP-NEAR.parquet")

LSTM_ADA = pd.read_parquet("../data/RMSPE/LSTM-ADA.parquet")
LSTM_AXS = pd.read_parquet("../data/RMSPE/LSTM-AXS.parquet")
LSTM_BTC = pd.read_parquet("../data/RMSPE/LSTM-BTC.parquet")
LSTM_DOGE = pd.read_parquet("../data/RMSPE/LSTM-DOGE.parquet")
LSTM_NEAR = pd.read_parquet("../data/RMSPE/LSTM-NEAR.parquet")

LSTM_BTC
```

```python
# Substitua esses DataFrames pelos seus dados
df1 = MLP_ADA * 100
df2 = MLP_AXS * 100
df3 = MLP_BTC * 100
df4 = MLP_DOGE * 100
df5 = MLP_NEAR * 100

# Cria uma lista com todos os dataframes
dfs = [df1, df2, df3, df4, df5]
labels = ["ADA", "AXS", "BTC", "DOGE", "NEAR"]

# Define a largura das barras
barWidth = 0.15

# Define a posição das barras
r1 = np.arange(len(df1.columns))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]

# Cria o gráfico
fig, ax = plt.subplots(figsize=(15, 8))

# Adiciona as barras
ax.bar(r1, df1.iloc[0], width=barWidth, label=labels[0])
ax.bar(r2, df2.iloc[0], width=barWidth, label=labels[1])
ax.bar(r3, df3.iloc[0], width=barWidth, label=labels[2])
ax.bar(r4, df4.iloc[0], width=barWidth, label=labels[3])
ax.bar(r5, df5.iloc[0], width=barWidth, label=labels[4])

# Adiciona as legendas
ax.set_xlabel('Modelos', fontweight='bold')
ax.set_ylabel('RMSPE (%)', fontweight='bold')
ax.set_xticks([r + barWidth for r in range(len(df1.columns))])
ax.set_xticklabels(df1.columns)
ax.set_title("RMSPE dos modelos MLP para diferentes criptomoedas")
ax.legend()

# Mostra o gráfico
plt.tight_layout()
plt.show()
```
